```python
def flip():
  import random
  import re
  print('Enter the number of coind to flip: ')
  nc = input()
  if re.match('[0-9]', nc):
    nc = int(nc)
    tf = 0
    results = []
    while tf < nc:
      r = random.randint(1, 2)
      if r == 1:
        results.append('heads')
        tf = tf + 1
      else:
        results.append('tails')
        tf = tf + 1
    count = results.count('heads')
    tcount = results.count('tails')
    print(results)
    return 'heads counter:', count, 'tails counter: ', tcount
  else:
    print('No numerical value try again MF(my friend)')
    flip()
flip()
```
